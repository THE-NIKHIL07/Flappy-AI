```python
import torch
from agent import DuelingDQN

# Load your trained model
n_actions = 2
model = DuelingDQN(n_actions)
model.load_state_dict(torch.load("models/best_model_new.pth"))
model.eval()

# Example input shape (batch=1, 7 features)
dummy_input = torch.randn(1, 7)
torch.onnx.export(model, dummy_input, "model/dueling_dqn_model.onnx",
                  input_names=["state"], output_names=["q_values"],
                  dynamic_axes={"state": {0: "batch"}, "q_values": {0: "batch"}})
```
